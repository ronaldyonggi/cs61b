## 2.3

Give the worst case and best case runtime in terms of `M` and `N`. Assume `ping` is in $\Theta(1)$ and returns an `int`.

```python
int j = 0;
for (int i = N; i > 0; i--) {
    for (; j <= M; j++) {
        if (ping (i, j) > 64) {
            break;
        }
    }
}
```

#### Answer

The `i` for loop is basically the same as,

```python
for (int i = 0; i < N; i++)
```

`break` means getting out of the current for loop, which is the current (`j`) loop. Thus we still need to loop through the `i` loop.

Notice that `j` is declared outside the loop! This means when `j` is incremented, `j` stays as that number.

Best case scenario is that we `break` early, and then for the rest of the loop, we don't need to worry about `j` iteration anymore.

Worst case scenario is the `j` iteration is huge.

* Best: $\Theta(N)$
* Worst: $\Theta(M+N)$

## 2.4

Give the worst case and best case runtime where `N` = `array.length`. Assume `mrpoolsort(array)` is in $\Theta(N log N)$ and returns `array` sorted.

```python
public static boolean mystery(int[] array) {
    array = mrpoolsort(array);
    int N = array.length;
    for (int i = 0; i < N; i += 1) {
        boolean x = false;
        for (int j = 0; j < N; j += 1) {
            if (i != j && array[i] == array[j]) {
                x = true;
            }
        }
        if (!x) {
            return false;
        }
    }
    return true;
}
```

#### (a)
What is `mystery()` doing?

#### Answer
Checks whether there's a duplicate in the input array
